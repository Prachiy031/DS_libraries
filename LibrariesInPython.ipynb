Correlation with heatmap   (correlation can only be done with floating value not with categorical data)
correlation ranges from -1 to 1

A correlation is the statistical summary of the relationship between two sets of variables. It is a core part of data exploratory analysis, and is a critical aspect of numerous advanced machine learning techniques.
Correlation explains how one or more variables are related to each other. These variables can be input data features which have been used to forecast our target variable.

Correlation, statistical technique which determines how one variables moves/changes in relation with the other variable. It gives us the idea about the degree of the relationship of the two variables. It’s a bi-variate analysis measure which describes the association between different variables. In most of the business it’s useful to express one subject in terms of its relationship with others.

A Heat map is a graphical representation of multivariate data that is structured as a matrix of columns and rows.

Heat maps are very useful in describing correlation among several numerical variables, visualizing patterns and anomalies.

A correlation matrix denotes the correlation coefficients between variables at the same time.

A heat map represents these coefficients to visualize the strength of correlation among variables. It helps find features that are best for Machine Learning model building.

The heat map transforms the correlation matrix into color coding.

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.columns
```

plotting data

initially create correlation matrix

```python
numeric_df = df.select_dtypes(include='number')   #as correlation matrix can only contain numeric data so select only numeric data
```

```python
corr_matrix=numeric_df.corr()
```

```python
sns.heatmap(corr_matrix)
```

customizing heatmaps

```python
plt.figure(1,figsize=(10,5))  #dimensions are in inch  1 is fig no
sns.heatmap(corr_matrix)
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG')   # cmap:maps data values to color space.   (basically it gives color as brown blue green) 
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG',center=0)   #center: specifies the value at which to center the colormap when plotting divergent data.
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG',center=0,annot=True)   #annot: when set to boolean True, displays the correlation coefficient for each matrix cell.
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG',center=0,annot=True,cbar=False)   #cbar: set to Boolean True by default. When set to False, it removes the color bar beside the heatmap.
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG',center=0,annot=True,cbar=False,linewidth=0.5,linecolor='white') 
```

```python
plt.figure(1, figsize=(10,5))
sns.heatmap(corr_matrix, cmap='BrBG', center=0, annot=True, xticklabels=False,yticklabels=False)
#yticklabels and xticklabels: control the presence of labels for the Y and X-axis respectively. They are set to Boolean True by default. When set to False, it removes the labels from the heatmap.
```

```python
plt.figure(1, figsize=(10,5))
sns.heatmap(corr_matrix, cmap='BrBG', center=0, annot=True, square=True)#square: when set to Boolean True, displays the heatmap in a squared form
```
